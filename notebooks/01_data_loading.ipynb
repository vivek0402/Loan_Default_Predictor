```python
import pandas as pd
```

```python
df = pd.read_csv(r"../data/accepted_2007_to_2018Q4.csv")
```

```python
df.head()
```

```python
print('Shape:', df.shape)
print('\nloan_status value counts:')
print(df['loan_status'].value_counts())

print('\nloan_status % distribution:')
print(df['loan_status'].value_counts(normalize=True).mul(100).round(2))
```

```python
# Drop ambiguous statuses
drop_statuses = ['Current', 'Late (31-120 days)', 'Late (16-30 days)', 
                 'In Grace Period', 'Default']

df = df[~df['loan_status'].isin(drop_statuses)]

# Map to binary target
status_map = {
    'Fully Paid': 0,
    'Does not meet the credit policy. Status:Fully Paid': 0,
    'Charged Off': 1,
    'Does not meet the credit policy. Status:Charged Off': 1
}

df['target'] = df['loan_status'].map(status_map)

print('Shape after filtering:', df.shape)
print('\nTarget distribution:')
print(df['target'].value_counts())
print('\nTarget % split:')
print(df['target'].value_counts(normalize=True).mul(100).round(2))
```

```python
# Fix mixed-type columns before saving
df = df.copy()  # also fixes PerformanceWarning
df['id'] = df['id'].astype(str)

# Save filtered dataset for next notebook
df.to_parquet('../data/loans_filtered.parquet', index=False)
print('Saved: loans_filtered.parquet')
print('Shape:', df.shape)
```
